```python
import pandas as pd
crime = pd.read_csv('data/SPD_Crime_Data__2008-Present_20240704.csv')
```

```python
crime.info()
```

```python

# Step 2: Group by the 'MCPP' column and count the number of offenses per neighborhood
offense_counts = crime.groupby(['MCPP', 'Offense Parent Group']).size().reset_index(name='Offense Count')

# Print the resulting DataFrame
print(offense_counts)

```

```python
crime['Offense Start DateTime'] = pd.to_datetime(crime['Offense Start DateTime'])

# Step 3: Filter the data for the year 2014
crime_2014 = crime[crime['Offense Start DateTime'].dt.year == 2014]

# Step 4: Group by the 'MCPP' and 'Offense Parent Group' columns and count the number of offenses per neighborhood and offense type
offense_counts_2014 = crime_2014.groupby(['MCPP', 'Offense Parent Group']).size().reset_index(name='Offense Count')
```

```python
offense_counts_2014.nunique()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming offense_counts_2014 is your original DataFrame
# Pivot the data for plotting
offense_pivot = offense_counts_2014.pivot_table(index='MCPP', columns='Offense Parent Group', values='Offense Count', aggfunc='sum').fillna(0)

# Calculate total offense counts for each neighborhood
offense_pivot['Total Offense Count'] = offense_pivot.sum(axis=1)

# Sort neighborhoods by total offense counts descending
offense_pivot_sorted = offense_pivot.sort_values(by='Total Offense Count', ascending=True)

# Plot the data
plt.figure(figsize=(14, 10))

offense_pivot_sorted.drop('Total Offense Count', axis=1).plot(kind='barh', stacked=True, figsize=(14, 10), cmap='viridis')

# Customize the plot
plt.title('Offense Counts per Neighborhood in 2014')
plt.xlabel('Offense Count')
plt.ylabel('Neighborhood')
plt.legend(title='Offense Type', bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the plot
plt.tight_layout()
plt.show()
```
